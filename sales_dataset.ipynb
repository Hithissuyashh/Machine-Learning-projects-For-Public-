```python
#Splitting up the data (importing the lib required for split data )
from sklearn.model_selection import train_test_split

```

```python
# (splitngh the target and feature variable )
x = data.drop("Salary", axis = 1)
y = data["Salary"]
```

```python
# split in train and test
x_train, x_test, y_train, y_test= train_test_split (x,y, test_size= 0.2, random_state= 42) #random state is used to set the data for the training since when we re start the data we want the model to read the exact data that was previous given for traning
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

```

```python
print("MAE is", mae) #how much error is occured that value is displayed
print("MSE is", mse) #penalty {square value of all the errors }
print("r2 is", r2) # if 1 = good, or closer to 1 is good if absolute 1 then the model will be in hardcode / over feed it will not perform the comparision model
```

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
dtr= DecisionTreeRegressor()
dtr.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr =  RandomForestRegressor (n_estimators=100, random_state=42)
rfr.fit(x_train, y_train)
y_pred_rfr=rfr.predict(x_test)
```

```python
mae=mean_absolute_error(y_test, y_pred_rfr)
mse=mean_squared_error(y_test, y_pred_rfr)
r2 = r2_score(y_test, y_pred_rfr)


print("MAE is", mae) #how much error is occured that value is displayed
print("MSE is", mse) #penalty {square value of all the errors }
print("r2 is", r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
xg =  GradientBoostingRegressor (n_estimators=100, random_state=42)
xg.fit(x_train, y_train)
y_pred_xg=xg.predict(x_test)
```

```python
mae=mean_absolute_error(y_test, y_pred_xg)
mse=mean_squared_error(y_test, y_pred_xg)
r2 = r2_score(y_test, y_pred_xg)


print("MAE is", mae) #how much error is occured that value is displayed
print("MSE is", mse) #penalty {square value of all the errors }
print("r2 is", r2)
```
